```python
# @title
import os
import shutil

input_directory = '/content/boltz_cif_files/'

# Create the directory if it doesn't exist
os.makedirs(input_directory, exist_ok=True)

# Move uploaded .cif files to the new directory
if 'uploaded' in locals() and uploaded:
    for filename in uploaded.keys():
        if filename.endswith('.cif'):
            source_path = os.path.join('/content/', filename)
            destination_path = os.path.join(input_directory, filename)
            try:
                shutil.move(source_path, destination_path)
                print(f"Moved '{filename}' to '{input_directory}'.")
            except FileNotFoundError:
                print(f"Warning: File '{filename}' not found. It might have been handled already or upload failed.")
        else:
            print(f"Skipping non-.cif file for moving: {filename}")
elif 'uploaded' not in locals():
    print("No files were uploaded. Please run the previous cell to upload files or continue to next step.")
else:
    print("No files were uploaded in the previous step.")
```

```python
import subprocess
import prody as pr
from pathlib import Path
import os
import shutil
from google.colab import files

input_directory = '/content/boltz_cif_files/'
output_directory = '/content/boltz_pdb_files/'

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)
print(f"Directory '{output_directory}' ensured to exist.")

# Perform CIF to PDB conversion and collect paths of generated PDBs
pdb_files_to_zip = []
for cif_path in Path(input_directory).glob('*.cif'):
    protein = pr.parseMMCIF(str(cif_path))
    pdb_filename = cif_path.with_suffix('.pdb')
    pr.writePDB(str(pdb_filename), protein)
    pdb_files_to_zip.append(pdb_filename)

# Copy generated PDB files to the output directory
for pdb_file_path in pdb_files_to_zip:
    if pdb_file_path.exists():
        destination_path = os.path.join(output_directory, pdb_file_path.name)
        shutil.copy(str(pdb_file_path), destination_path)
        print(f"Copied '{pdb_file_path.name}' to '{output_directory}'.")

# Zip the output directory with -j flag to flatten directory structure
zip_filename = 'boltz_pdb_files.zip'
# Collect all pdb files from the output directory to include them individually
pdb_files_in_output_dir = [str(p) for p in Path(output_directory).glob('*.pdb')]

if pdb_files_in_output_dir:
    zip_command = ['zip', '-j', zip_filename] + pdb_files_in_output_dir
    subprocess.run(zip_command, check=True)
    print(f"Zipped PDB files into '{zip_filename}' using -j flag.")
else:
    print("No PDB files found in the output directory to zip.")

# Download the zipped file
files.download(zip_filename)
```
